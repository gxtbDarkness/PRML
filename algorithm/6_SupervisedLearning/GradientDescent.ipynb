# 梯度下降

本文件用于实现批梯度下降与随机梯度下降相关的内容

```python
import numpy as np
import matplotlib.pyplot as plt
```

### 批梯度下降

```python
R1 = np.eye(3, dtype=float) - 0.04 * np.array([[25, -15, -20], 
                                               [-15, 9, 12], 
                                               [-20, 12, 16]])
A = np.array([[0, -20, -14],
             [3, 27, -4],
             [4, 11, -2]])

print(R1 @ A)
```
